# Seaborn - Styles and Colors

* Topics:
    * Background
    * Ticks
    * Limits
    * Size and Aspect
    * Scale and Context

```python
# Import Libraries

import numpy as np
from numpy.random import randn
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy import stats

# Import Data

tips = sns.load_dataset('tips')
```

```python
# Style: Background

sns.set_style('white')
sns.countplot(data=tips,x='sex')
```

```python
# Style: Ticks

sns.set_style('ticks')
sns.countplot(data=tips,x='sex',palette='deep')
```

```python
# Style: Limits Removal

sns.countplot(data=tips, x='sex')
sns.despine(left=True)
```

```python
# Size and Aspect

plt.figure(figsize=(12,3))
sns.countplot(data=tips,x='sex')
```

```python
# Size and Aspect

sns.lmplot(data=tips, x='total_bill', y='tip', size=2, aspect=5)
```

```python
# Scale and Context

sns.set_context('poster',font_scale=3)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```
