```python
# Importar Bibliotecas Auxiliares

import pandas as pd
import numpy as np
from numpy.random import randn

# Importar Biblioteca - Python Plot: matplotlib.plotly

import matplotlib.pyplot as plt

# "No Jupyter": Para visualizar resultados no notebook, e não precisar da função "plt.show()"

%matplotlib inline
```

```python
# Gerar figura e eixos simultaneamente
# (sem precisar utilizar "plt.figure()" e "plt.axes()")

fig, axes = plt.subplots()

# Dados

x = np.linspace(0, 5, 11)
y = x ** 2

# Utilizar o objeto dos eixos para adicionar informações ao gráfico

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Título');
```

```python
# Indicar a quantidade de subgráficos (axes)

fig, axes = plt.subplots(nrows=1,ncols=2)

# "axes" é um array com os subgraficos de uma figura.

for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título')

# Melhorar Visualização (nenhum elemento cobrir outro dentro da mesma figura)

fig.tight_layout() # poderia utilizar: plt.tight_layout()
```

```python
# Figure Size and Dots Per Inch (DPI)

# figsize: é uma tupla da largura e altura da figura em polegadas
# dpi: quantidade de pontos por polegada

fig, axes = plt.subplots(figsize=(8,2),dpi=100)
axes.plot(x,y,'r')

# Figure save

# Supported Formats: 'eps' 'jpeg' 'jpg''pdf' 'pgf' 'png' 'ps' 'raw' 'rgba' 'svg' 'svgz' 'tif' 'tiff'

fig.savefig("MatplotLib_01_general.png", dpi=200)
```

```python
# Data

x = np.linspace(0, 5, 11)
y1 = x ** 2
y2 = x ** 3

# Figure Attributes

fig, ax = plt.subplots(figsize=(8,2),dpi=100)   # figure: fig = plt.figure()   
                                                # axes:   ax = fig.add_axes([0,0,1,1])

ax.set_title('title')                           # title => plt.title('title')
ax.set_xlabel("x")                              # xlabel =>  plt.xlabel('x')
ax.set_ylabel("y")                              # ylabel =>  plt.ylabel('y')

ax.plot(x,y1,'r-',label="x**2");                # plot figure => plt.plot()
ax.plot(x,y2,'g--',label="x**3");               # Colors and lines: same as matlab style

ax.grid()                                       # Activate grid
ax.set_xticks(range(0,6,1))                     # x divisions of grids
ax.set_yticks(range(0,140,20))                  # y divisions of grids

ax.axis('tight')                                # Options: on off equal scaled tight auto normal image square
ax.set_xlim([-0.1,5.1])                         # overwrite axis command. plt.xlim()
ax.set_ylim([-10,140])                          # overwrite axis command. plt.ylim()

ax.legend(loc=0)                                # legend -> plt.legend(['First', 'Second'])
                                                # loc = 0, 1, 2, 3, 4 (BEST, HR, HL, LL, LR)

ax.annotate(xy=[4,1], s = 'Annotation!');       # annotations -> plt.annotate()
```

```python
# Colors Definition

x = np.linspace(0,5,11)

fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # 0.5 of transparency
ax.plot(x, x+2, color="#8B008B");        # RGB
```

```python
# Line shape and Markers

# lw = line width
# ls = line shape = '-','--','-.',':','steps'
# marker = '+','o','*','s',',','.','1','2','3', ...
# marker size, markerfacecolor, markeredgewidth, markeredgecolor

fig, ax = plt.subplots(figsize=(12,6))

# Different Widths

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# Different Line Shapes

ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# Stylized Lines

line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # Format: line length, space lenght...

# Line Shapes and Markers

ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# Marker size and color

ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
