# Seaborn - Regression and PairGrid

* Topics:
    * Linear Models 1 (Markers)
    * Linear Models 2 (Grid of lm)
    * Pair Grid       (different types of plots in the same grid)
    * PairPlot        (pdf in diagonal and joint distribution in others positions)
    * FacetGrid       (Pair Grid based on a resource)
    * JointGrid       (Grid version of jointplot() )

```python
# Import Libraries

import numpy as np
from numpy.random import randn
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy import stats

# Import Data

tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
# Linear Model 1: Markers

# markers => https://matplotlib.org/api/markers_api.html
# scatter_kws={'s':20} => defines size of markers

sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',markers=['o','v'],palette='rainbow',scatter_kws={'s':20})
```

```python
# Linear Models 2: Grids

# could use "lin" parameter

sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',col='day',aspect=0.8,size=8)
```

```python
# PairGrid 1: Just one Type

g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# PairGrid 2: More than one type

g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
# Pairplot

sns.pairplot(iris,hue='species',palette='rainbow')
```

```python
# FacetGrid

g = sns.FacetGrid(tips, col="time", row="smoker")
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

```python
# Joint Grid

g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```
