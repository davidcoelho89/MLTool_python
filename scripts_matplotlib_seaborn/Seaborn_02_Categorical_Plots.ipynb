# Seaborn - Categorical Plots

* TÃ³picos:
    * barplot    (aggregate categorized data based on some function)
    * countplot  (count the number of occurrences)
    * boxplot    (compare categorial data. Shows "quartis" and "outliers")
    * violinplot (shows boxplot and kde at the same graphic)
    * stripplot  (scatter of one variable, and one categorical)
    * swarmplot  (a strip plot without overlapping data)
    * factorplot (general categorical plot. "kind" parameter)
    * Merge Categorial Plots

```python
# Import Libraries

import numpy as np
from numpy.random import randn
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy import stats

# Import Data

tips = sns.load_dataset('tips')
tips.head()
```

```python
# Bar Plot

# Estimators: np.std / np.mean (default)

sns.barplot(data=tips, x='sex', y='total_bill', estimator=np.mean)
```

```python
# Count Plot

sns.countplot(data=tips, x='sex')
```

```python
# Boxplot 1: vertical

# hue = for each value of this variable, a different color 
# pallete = color pallete

sns.boxplot(data=tips, x="day", y="total_bill", hue='sex', palette='rainbow')
```

```python
# Boxplot 2: horizontal

sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
# Violin Plot

sns.violinplot(data=tips, x="day", y="total_bill", hue = 'sex', split = True, palette='rainbow')
```

```python
# Strip Plot

# jitter = True (points horizontally separated)
# hue = one color for each category
# dodge = True (separate data by "hue" variable)

sns.stripplot(data=tips, x='day', y='total_bill', jitter=True, hue='sex', dodge=True, palette='Set1')
```

```python
# Swarmplot

sns.swarmplot(data=tips, x="day", y="total_bill", hue='sex', dodge=True, palette="coolwarm")
```

```python
# Factor Plot

# kind = type of plot ('bar','count','box','violin','strip','swarm')

sns.factorplot(data=tips, x='sex', y='total_bill', kind='bar')
```

```python
# Merge Categorical Plots

sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```
