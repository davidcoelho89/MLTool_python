```python
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Import DataFrames

df1 = pd.read_csv('Pandas_07_Visualize_df1',index_col=0)
df2 = pd.read_csv('Pandas_07_Visualize_df2')

# Information About DataFrame

print(df1.head())
print()

print(df1.info())
print()
```

```python
# Visualize Histograms (default / without style)

df1['A'].hist()
```

```python
# bmh

plt.style.use('bmh')
df1['A'].hist()
```

```python
# Dark Background

plt.style.use('dark_background')
df1['A'].hist()
```

```python
# 538

plt.style.use('fivethirtyeight')
df1['A'].hist()
```

```python
# gg_plot

plt.style.use('ggplot')
df1['A'].hist()
```

```python
# Area

df2.plot.area(alpha=0.4)
```

```python
# Bar Plots 1

df2.plot.bar()
```

```python
# Bar Plots 2

df2.plot.bar(stacked=True)
```

```python
# Histograms

df1['A'].plot.hist(bins=50)
```

```python
# line Plots

df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1)
```

```python
# Scatter Plots 1

df1.plot.scatter(x='A',y='B')
```

```python
# Scatter Plots 2

# Color of points controlled by C.
# The color could be, for example, the value of a function [C = f(A,B)]

df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
# Scatter Plots - 3

# Size of points controlled by C

df1.plot.scatter(x='A',y='B',s = df1['C']*9)
```

```python
# Boxplots

df2.plot.box()
```

```python
# Hexagonal Bin Lote

df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

```python
# Kernel Density Estimation - KDE 1

df2['a'].plot.kde()
```

```python
# Kernel Density Estimation - KDE 2

df2.plot.density()
```
