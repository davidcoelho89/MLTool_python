```python
# Importar Bibliotecas Auxiliares

import pandas as pd
import numpy as np
from numpy.random import randn

# Importar Biblioteca - Python Plot: matplotlib.plotly

import matplotlib.pyplot as plt

# "No Jupyter": Para visualizar resultados no notebook, e não precisar da função "plt.show()"

%matplotlib inline

# Verificar Plots interativos

# %matplotlib notebook
```

```python
# Comandos Basicos

x = np.linspace(0, 5, 11)
y = x ** 2

# Gerar Grafico (Matlab-style)

plt.plot(x, y, 'r') 
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
plt.legend(['teste'])
plt.show()
```

```python
# Subplots (Matlab-Style)

plt.subplot(1,2,1)      # (num_lin,num_col,which_line)
plt.plot(x,y,'r--') 
plt.subplot(1,2,2)      # (num_lin,num_col,which_line)
plt.plot(y,x,'g*-');
```

```python
# MatplotLib Orientado a Objetos (criar instancia)

fig = plt.figure()

# Adicionar Eixo (esquerda, inferior, largura, altura -> faixa de 0 a 1)

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
                    # LB, VB,  %H,  %V
    
# Plotar figura no eixo indicado (cada figura tem um ou mais eixos)

axes.plot(x,y,'b')
axes.set_xlabel('Define o Label X') 
axes.set_ylabel('Define o Label Y')
axes.set_title('Define o Título');
```

```python
# Mais de um eixo na mesma figura

fig = plt.figure()

# Eixo maior

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
axes1.plot(x,y,'b')
axes1.set_xlabel('Eixo X1')
axes1.set_ylabel('Eixo Y1')
axes1.set_title('Título 1')

# Eixo menor

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) 
axes2.plot(y,x,'r')
axes2.set_xlabel('Eixo X2')
axes2.set_ylabel('Eixo Y2')
axes2.set_title('Título 2');
```
